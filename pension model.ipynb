```python
"""
Pension Model - Eden Shrian
"""

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
raise_in_wage = 1.0124
yield_from_savings = 1.031
pension_rate = 0.185
interest_rate = 1.021
avg_age_death_women = 84
initial_age = 25
num_of_years = 40

def calc_monthly_effect(rate):
    return (rate) ** (1/12)

class pension:
    
    pass

    def __init__(self,wage,pension_rate,interest_rate):
        
        self.wage = wage
        self.interest_rate = [1/(1+interest_rate)]
        self.pension_saving = wage*pension_rate

    def filling_lists(self,num_of_years,raise_wage,wage): 
        
        num_of_years = num_of_years*12
        
        self.wage = [wage*calc_monthly_effect(raise_wage)**i for i in range(num_of_years)]
        self.interest_rate = [1/(calc_monthly_effect(interest_rate))**i for i in range(num_of_years)]
        self.pension_saving = [wage*(calc_monthly_effect(raise_wage**i))*pension_rate*calc_monthly_effect(yield_from_savings) for i in range(num_of_years)]
        self.pv = [wage*(calc_monthly_effect(raise_wage)**i)*pension_rate**calc_monthly_effect(yield_from_savings)*1/(1+calc_monthly_effect(interest_rate))**i for i in range(num_of_years)]
        
        return self.wage,self.interest_rate,self.pension_saving
    
    def create_dataframe(self):
        
        dictionary = dict({
            'wage': self.wage,
            'interest_rate': self.interest_rate,
            'pension_saving' : self.pension_saving,
            'pv': self.pv
        })
        df = pd.DataFrame.from_dict(dictionary)
        return df
    
    def calc_cumulative_provisions(self):
        return round(sum(self.pension_saving),2)
    
    def calc_income_for_retirenment(self):
        
        age_of_retirenement = initial_age +  num_of_years
        average_time_to_live = avg_age_death_women - age_of_retirenement
        average_time_to_live = average_time_to_live*12
        
        return round(sum(self.pension_saving)/average_time_to_live,2)
```
